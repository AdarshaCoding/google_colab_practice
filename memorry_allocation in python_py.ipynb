```python
from sys import getsizeof
L = []
for i in range(18):
  L.append(i*10)
  print(L, end=" - size: ")
  print(getsizeof(L))
  print(id(L))

print(L)
address = []
for el in L:
  address.append(id(el))

for i in range(len(address)-1):
  print(address[i], address[i+1], end= " : ")
  print(abs(address[i] - address[i+1]))
```
